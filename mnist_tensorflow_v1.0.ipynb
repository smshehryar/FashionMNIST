```python
import numpy as np;
import tensorflow as tf;
from tensorflow.examples.tutorials.mnist import input_data;

mnist = input_data.read_data_sets("MNIST_data/", one_hot=True);
```

```python
# >>> Outline the Model

inputSize = 784;
outputSize = 10;
hiddenSize = 200;


tf.reset_default_graph();

# set the placeholders
inputs = tf.placeholder(tf.float32, [None, inputSize]);
targets = tf.placeholder(tf.float32, [None, outputSize]);

# Declare variables for input and hidden_1 layers

weights_1 = tf.get_variable("weights_1", [inputSize, hiddenSize]);
biases_1 = tf.get_variable("biases_1", [hiddenSize]);

# Applying activation function
outputs_1 = tf.nn.relu(tf.matmul(inputs, weights_1)+ biases_1);


# Declare variables for hidden_1 and hidden_2 layers

weights_2 = tf.get_variable("weights_2",[hiddenSize,hiddenSize]);
biases_2 = tf.get_variable("biases_2", [hiddenSize]);

outputs_2 = tf.nn.relu(tf.matmul(outputs_1,weights_2)+biases_2);

# Declare variables for hidden_2 and Output layers

weights_3 = tf.get_variable("weights_3", [hiddenSize,outputSize]);
biases_3 = tf.get_variable("biases_3", [outputSize]);

outputs = tf.matmul(outputs_2,weights_3)+biases_3;

# Apply the softmax activation function along with entropy loss function.
loss = tf.nn.softmax_cross_entropy_with_logits(logits=outputs, labels=targets);

meanLoss = tf.reduce_mean(loss);


optimize = tf.train.AdamOptimizer(learning_rate=0.001).minimize(meanLoss);


# out_equal_targets - representing a match or not
out_equals_targets = tf.equal(tf.argmax(outputs, 1), tf.argmax(targets, 1));

# sum up 'out_equal_targets' and calculate mean
accuracy = tf.reduce_mean(tf.cast(out_equals_targets, tf.float32));

# running session and initializing variables
sess = tf.InteractiveSession();
initializer = tf.global_variables_initializer();
sess.run(initializer);

# just an intuitively reasonable size, although it is a hyperparameter and should be tweaked later on
batch_size = 100;

batches_number = mnist.train._num_examples // batch_size;

max_epochs = 15;

prev_validation_loss = 9999999.


for epoch_counter in range(max_epochs):
    
    current_epoch_loss = 0.
    
    for batch_counter in range(batches_number):
        
        input_batch, target_batch = mnist.train.next_batch(batch_size);
        
        _, batch_loss = sess.run([optimize, meanLoss],
                                 feed_dict = {inputs: input_batch, targets: target_batch});
        
        current_epoch_loss += batch_loss;
        
        
    current_epoch_loss /= batches_number;
    
    input_batch, target_batch = mnist.validation.next_batch(mnist.validation._num_examples);
    
    validation_loss, validation_accuracy = sess.run([meanLoss, accuracy], 
                                                    feed_dict={inputs: input_batch, targets: target_batch});
    
    
    print('Epoch '+str(epoch_counter+1)+
          '. Training Loss: '+'{0:.03f}'.format(current_epoch_loss)+
          '. Validation Loss: '+'{0:.03f}'.format(validation_loss)+
          '. Validation Accuracy: '+'{0:.02f}'.format(validation_accuracy*100.)+'%');
    
    
    if validation_loss > prev_validation_loss:
        break
        
    prev_validation_loss = validation_loss
    
print('Training Complete!');



input_batch, target_batch = mnist.test.next_batch(mnist.test._num_examples);
test_accuracy = sess.run([accuracy],
                        feed_dict={inputs: input_batch, targets: target_batch});

test_accuracy_percent = test_accuracy[0]*100;

print('Test Data Accuracy: '+'{0:.02f}'.format(test_accuracy_percent)+'%');
```
